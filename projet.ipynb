```python
import pandas as pd
import matplotlib.pyplot as plt

# Charger le fichier CSV dans un DataFrame
df = pd.read_csv("DonneesCollectees/caracteristics.csv")

# Extraire l'heure de la journée à partir de la colonne 'hrmn'
df['heure'] = df['hrmn'] // 100

# Compter le nombre d'accidents pour chaque heure de la journée
accidents_par_heure = df['heure'].value_counts().sort_index()

# Calculer le pourcentage d'accidents pour chaque heure de la journée
pourcentage_accidents_par_heure = (accidents_par_heure / accidents_par_heure.sum()) * 100

# Créer le diagramme circulaire
plt.figure(figsize=(8, 8))
plt.pie(pourcentage_accidents_par_heure, labels=pourcentage_accidents_par_heure.index, autopct='%1.1f%%', startangle=140)
plt.title('Pourcentage d\'accidents par heure de la journée')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
